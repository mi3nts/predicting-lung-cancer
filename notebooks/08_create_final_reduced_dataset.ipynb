# 08 - Create Final Reduced Dataset

## Purpose
This notebook creates the final reduced dataset by:
1. **Dropping redundant demographics variables** (5 variables identified in notebook 06)
2. **Dropping irrelevant weather variables** (58 variables from notebook 07 step 4.3)
3. **Dropping redundant weather variables** (21 variables from correlation analysis in notebook 07)
4. **Formatting weather variable names** to Title Case for publication
5. **Verifying data quality** and final feature set
6. **Saving final reduced dataset** ready for modeling

## Input
- `data/combined_final/final_combined_all_variables.csv` (from notebook 05)

## Output
- `data/combined_final/final_combined_all_variables_reduced.csv`


## 1. Import Libraries

```python
import pandas as pd
import numpy as np
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

print(f"pandas version: {pd.__version__}")
```

## 2. Load Combined Dataset

```python
df = pd.read_csv('../data/combined_final/final_combined_all_variables.csv')

print("=" * 70)
print("ORIGINAL DATASET LOADED")
print("=" * 70)
print(f"Shape: {df.shape}")
print(f"  - Rows: {df.shape[0]:,}")
print(f"  - Columns: {df.shape[1]}")
print(f"  - Years: {sorted(df['Year'].unique())}")

original_rows = df.shape[0]
original_cols = df.shape[1]
```

## 3. Define Variables to Drop

### 3.1 Redundant Demographics Variables (from notebook 06)

```python
demographics_to_drop = [
    'Rent Burden (+50% of HI)',
    'Median Household Income',
    'Gini Index',
    'High School Degree or Higher (%)',
    'Median Age'
]

print(f"Demographics variables to drop: {len(demographics_to_drop)}")
for i, var in enumerate(demographics_to_drop, 1):
    print(f"  {i}. {var}")

print("\nRetained behavioral variable:")
print("  - Smoking Rate")
```

### 3.2 Irrelevant Weather Variables (from notebook 07 step 4.3)

```python
# Irrelevant weather variables (from notebook 07 step 4.3)
irrelevant_weather = [
    'PM$_{10}$',
    'FoT PM$_{10}$ above EPA threshold',
    'FoT PM$_{10}$ above75ᵗʰ percentile',
    'FoT PM$_{1}$ above75ᵗʰ percentile',
    'PM$_1$',
    'Specific humidity.1',
    'Total column water',
    '2m dew point temperature',
    '2m temperature',
    'PM$_{2.5}$',
    'FoT Temperature above 90 $\\mathrm{^o F}$',
    'FoT Temperature below 0 $\\mathrm{^o C}$',
    'FoT PM$_{2.5}$ above EPA threshold',
    'Temperature',
    'Surface geopotential',
    'Dust aerosol (0.03-0.55 µm) mixing ratio',
    'Nitrogen dioxide',
    'Peroxyacetyl nitrate',
    'Formaldehyde',
    'Sulphur dioxide',
    'Nitric acid',
    'Propane',
    'Nitrogen monoxide',
    'Ozone',
    'Hydrogen peroxide',
    'Specific humidity',

    # Total AOD measurements (redundant aerosol measurements)
    'Total AOD at 469 nm',
    'Total AOD at 550 nm',
    'Total AOD at 670 nm',
    'Total AOD at 865 nm',
    'Total AOD at 1240 nm',

    # Total column gas measurements (redundant with other air quality measures)
    'Total column carbon monoxide',
    'Total column ethane',
    'Total column formaldehyde',
    'Total column hydrogen peroxide',
    'Total column hydroxyl radical',
    'Total column isoprene',
    'Total column methane',
    'Total column nitric acid',
    'Total column nitrogen dioxide',
    'Total column ozone',
    'Total column peroxyacetyl nitrate',
    'Total column propane',
    'Total column sulphur dioxide',
    'Total column water vapour',
    'Total column nitrogen monoxide',

    # Vertically integrated aerosol masses (redundant)
    'Vertically integrated mass of dust aerosol (0.03-0.55 µm)',
    'Vertically integrated mass of dust aerosol (0.55-9 µm)',
    'Vertically integrated mass of dust aerosol (9-20 µm)',
    'Vertically integrated mass of hydrophilic black carbon aerosol',
    'Vertically integrated mass of hydrophilic organic matter aerosol',
    'Vertically integrated mass of hydrophobic organic matter aerosol',
    'Vertically integrated mass of hydrophobic black carbon aerosol',
    'Vertically integrated mass of sea salt aerosol (0.03-0.5 µm)',
    'Vertically integrated mass of sea salt aerosol (0.5-5 µm)',
    'Vertically integrated mass of sea salt aerosol (5-20 µm)',
    'Vertically integrated mass of sulphate aerosol',
    'Vertically integrated mass of sulphur dioxide'
]

print(f"Irrelevant weather variables specified: {len(irrelevant_weather)}")
for i, var in enumerate(irrelevant_weather[:10], 1):
    print(f"  {i:2}. {var}")
print("  ...")
```

### 3.3 Redundant Weather Variables (from notebook 07 correlation analysis)


```python
# Redundant weather variables (from notebook 07 correlation analysis)
redundant_weather = [
    'Black carbon AOD at 550 nm',
    'Sulphate AOD at 550 nm',
    'Dust AOD at 550 nm',
    'Sea salt AOD at 550 nm',
    'Organic matter AOD at 550 nm',
    'Hydrophilic organic matter aerosol mixing ratio',
    'Sea salt aerosol (0.03-0.5 µm) mixing ratio',
    'Hydroxyl radical',
    'FoT Temperature below25ᵗʰ percentile',
    'FoT Temperature above75ᵗʰ percentile',
    'Sea salt aerosol (0.03-0.5 µm) mixing ratio',
    'Carbon monoxide',
    'Ethane',
    'Isoprene',
    'Snow albedo',
    'Surface pressure',
    'FoT Peroxyacetyl Nitrate above75ᵗʰ percentile',
    'FoT Isoprene above75ᵗʰ percentile',
    'FoT Hydrogen peroxide above75ᵗʰ percentile',
    'Relative humidity',
    'Buffalo'    
]

print(f"Redundant weather variables to drop: {len(redundant_weather)}")
for i, var in enumerate(redundant_weather, 1):
    print(f"  {i:2}. {var}")
```

### 3.4 Combine All Variables to Drop

```python
all_vars_to_drop = demographics_to_drop + irrelevant_weather + redundant_weather

print("=" * 70)
print("SUMMARY: VARIABLES TO DROP")
print("=" * 70)
print(f"  Demographics:       {len(demographics_to_drop)}")
print(f"  Irrelevant weather: {len(irrelevant_weather)}")
print(f"  Redundant weather:  {len(redundant_weather)}")
print(f"  TOTAL:              {len(all_vars_to_drop)}")
```

## 4. Check Which Variables Exist

```python
cols_to_drop = [col for col in all_vars_to_drop if col in df.columns]
cols_not_found = [col for col in all_vars_to_drop if col not in df.columns]

print("=" * 70)
print("COLUMNS TO DROP (THAT EXIST IN DATASET)")
print("=" * 70)
print(f"Specified:        {len(all_vars_to_drop)}")
print(f"Found in dataset: {len(cols_to_drop)}")
print(f"Not found:        {len(cols_not_found)}")

if cols_not_found:
    print(f"\nColumns NOT found (already absent from dataset):")
    for col in cols_not_found:
        print(f"  - {col}")
```

## 5. Drop Redundant Columns

```python
print("=" * 70)
print("DROPPING REDUNDANT COLUMNS")
print("=" * 70)
print(f"\nOriginal shape: {df.shape}")

df_reduced = df.drop(columns=cols_to_drop)

reduced_rows = df_reduced.shape[0]
reduced_cols = df_reduced.shape[1]
cols_dropped = original_cols - reduced_cols

print(f"Reduced shape:  {df_reduced.shape}")
print(f"  - Rows:            {reduced_rows:,} (no change)")
print(f"  - Columns:         {reduced_cols}")
print(f"  - Columns dropped: {cols_dropped}")
print(f"  - Retention:       {(reduced_cols / original_cols) * 100:.1f}%")
```

## 6. Identify Remaining Variables

```python
print("=" * 70)
print(f"REMAINING COLUMNS ({len(df_reduced.columns)})")
print("=" * 70)

identifiers = ['County', 'State', 'Year', 'fips']
target = ['Lung Cancer Mortality Rate']

remaining_demographics = [
    'Total Population', 'Poverty Rate', 'Unemployment Rate', 'Disability Rate',
    "Bachelor's Degree or Higher (%)", 'White Population (%)',
    'Hispanic Population (%)', 'Black Population (%)',
    'Households with No Vehicle (%)', 'Single Mother Families (%)',
    'Smoking Rate'
]
remaining_demographics = [c for c in remaining_demographics if c in df_reduced.columns]

control_cols = ['is_post_2015']
control_remaining = [c for c in control_cols if c in df_reduced.columns]

livestock_cols = ['Buffalo', 'Cattle', 'Chicken', 'Duck', 'Goat', 'Horse', 'Pig', 'Sheep']
livestock_remaining = [c for c in livestock_cols if c in df_reduced.columns]

other_cols = [c for c in df_reduced.columns
              if c not in identifiers + target + remaining_demographics + control_remaining + livestock_remaining]

print(f"\nIdentifiers ({len([c for c in identifiers if c in df_reduced.columns])}):"
      f" {[c for c in identifiers if c in df_reduced.columns]}")
print(f"Target (1): {target}")
print(f"\nDemographics/behavioral retained ({len(remaining_demographics)}):")
for col in remaining_demographics:
    print(f"  - {col}")
print(f"\nControl variables retained ({len(control_remaining)}): {control_remaining}")
print(f"\nLivestock retained ({len(livestock_remaining)}):")
for col in livestock_remaining:
    print(f"  - {col}")
print(f"\nWeather variables retained ({len(other_cols)}):")
for col in other_cols:
    print(f"  - {col}")
```

## 7. Format Weather Variable Names to Title Case

```python
def to_title_case_preserve_special(name):
    words = name.split(' ')
    titled_words = []
    for word in words:
        if word and 'a' <= word[0] <= 'z':
            titled_words.append(word[0].upper() + word[1:])
        else:
            titled_words.append(word)
    return ' '.join(titled_words)

def normalize_unit_symbols(name):
    return name.replace('Μm', 'µm')

# Only rename weather variables (not identifiers, target, demographics, controls, or livestock)
non_weather = (identifiers + target + remaining_demographics +
               control_remaining + livestock_remaining + ['Fips'])

weather_name_mapping = {}
for col in df_reduced.columns:
    if col not in non_weather:
        new_name = normalize_unit_symbols(to_title_case_preserve_special(col))
        if new_name != col:
            weather_name_mapping[col] = new_name

print(f"Weather variable renames: {len(weather_name_mapping)}")
for old, new in weather_name_mapping.items():
    print(f"  '{old}' -> '{new}'")
```

```python
df_reduced = df_reduced.rename(columns=weather_name_mapping)
df_reduced = df_reduced.rename(columns=lambda col: col.replace('Μm', 'µm'))
print(f"Renamed {len(weather_name_mapping)} weather variables to Title Case")
```

## 8. Verification

### 8.1 Missing Values

```python
print("=" * 70)
print("DATA QUALITY CHECK")
print("=" * 70)

missing_values = df_reduced.isnull().sum()
total_missing = missing_values.sum()
print(f"\nMissing values: {total_missing}")
if total_missing > 0:
    for col in df_reduced.columns:
        if missing_values[col] > 0:
            pct = (missing_values[col] / len(df_reduced)) * 100
            print(f"  {col}: {missing_values[col]:,} ({pct:.2f}%)")
else:
    print("  No missing values")
```

### 8.2 Year Distribution

```python
print("\nObservations per year:")
year_counts = df_reduced['Year'].value_counts().sort_index()
for year, count in year_counts.items():
    print(f"  {year}: {count:,} counties")
print(f"\nAverage counties per year: {year_counts.mean():.0f}")
```

### 8.3 Target Variable

```python
print("\nTarget variable statistics (Lung Cancer Mortality Rate):")
print(f"  Count:  {df_reduced['Lung Cancer Mortality Rate'].count():,}")
print(f"  Min:    {df_reduced['Lung Cancer Mortality Rate'].min():.4f}")
print(f"  Max:    {df_reduced['Lung Cancer Mortality Rate'].max():.4f}")
print(f"  Mean:   {df_reduced['Lung Cancer Mortality Rate'].mean():.4f}")
print(f"  Median: {df_reduced['Lung Cancer Mortality Rate'].median():.4f}")
print(f"  Std:    {df_reduced['Lung Cancer Mortality Rate'].std():.4f}")
```

### 8.4 Final Column List

```python
print("=" * 70)
print(f"FINAL COLUMN LIST ({len(df_reduced.columns)} columns)")
print("=" * 70)
for i, col in enumerate(df_reduced.columns, 1):
    print(f"  {i:2}. {col}")
```

## 9. Save Final Reduced Dataset

```python
output_dir = Path('../data/combined_final')
output_dir.mkdir(parents=True, exist_ok=True)

output_path = output_dir / 'final_combined_all_variables_reduced.csv'
df_reduced.to_csv(output_path, index=False)

print("=" * 70)
print("FINAL REDUCED DATASET SAVED")
print("=" * 70)
print(f"File: {output_path}")
print(f"  - Shape:    {df_reduced.shape}")
print(f"  - Rows:     {df_reduced.shape[0]:,}")
print(f"  - Columns:  {df_reduced.shape[1]}")
print(f"  - File size: {output_path.stat().st_size / (1024*1024):.2f} MB")
```

```python
print("=" * 70)
print("FEATURE REDUCTION SUMMARY")
print("=" * 70)
print(f"\n1. CHANGES MADE:")
print(f"   Dropped {cols_dropped} redundant variables")
print(f"     - Demographics:       {len([c for c in demographics_to_drop if c in df.columns])}")
print(f"     - Irrelevant weather: {len([c for c in irrelevant_weather if c in df.columns])}")
print(f"     - Redundant weather:  {len([c for c in redundant_weather if c in df.columns])}")
print(f"   Formatted {len(weather_name_mapping)} weather variable names to Title Case")

print(f"\n2. DATASET COMPARISON:")
print(f"   Original: {original_cols} columns, {original_rows:,} rows")
print(f"   Reduced:  {reduced_cols} columns, {reduced_rows:,} rows")

print(f"\n3. READY FOR MODELING:")
print(f"   Input file for notebook 09:")
print(f"   data/combined_final/final_combined_all_variables_reduced.csv")
```
